# 1512. Number of Good Pairs

Given an array of integers `nums`, return the number of good pairs.

A pair `(i, j)` is called good if `nums[i] == nums[j]` and `i < j`.

### Examples
**Example 1:**
* **Input:** `nums = [1,2,3,1,1,3]`
* **Output:** `4`
* **Explanation:** There are 4 good pairs (0,3), (0,4), (3,4), (2,5) 0-indexed.

**Example 2:**
* **Input:** `nums = [1,1,1,1]`
* **Output:** `6`
* **Explanation:** Each pair in the array are good.

**Example 3:**
* **Input:** `nums = [1,2,3]`
* **Output:** `0`

## 1. Initial Approach: Array / Enumeration $O(N^2)$
The initial approach checks every possible pair `(i, j)` where `i < j`. While easy to read, it is inefficient for large inputs because of the nested loops.

```python
from typing import List

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        length = len(nums)
        num_good_pairs = 0

        for i in range(length):
            for j in range(i+1, length):
                if nums[i] == nums[j] and i < j:
                    num_good_pairs += 1
        
        return num_good_pairs

# Test Cases
sol = Solution()
print(f"Test 1: {sol.numIdenticalPairs([1,2,3,1,1,3])}")  # Expected: 4
print(f"Test 2: {sol.numIdenticalPairs([1,1,1,1])}")      # Expected: 6
print(f"Test 3: {sol.numIdenticalPairs([1,2,3])}")        # Expected: 0
```

## 2. Optimized Approach: Hash Table Single-Pass $O(N)$
Instead of looking ahead with a nested loop, we keep a running count of how many times we've seen each number. Every time we encounter a number we've seen before, it forms a new "good pair" with *every previous occurrence* of that number.

```python
class SolutionOptimized:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        counts = {}
        num_good_pairs = 0

        for num in nums:
            if num in counts:
                # Add the number of times we've seen this 'num' before
                num_good_pairs += counts[num]
                counts[num] += 1
            else:
                # First time seeing this number
                counts[num] = 1
                
        return num_good_pairs

# Test Cases
sol_opt = SolutionOptimized()
print(f"Test 1: {sol_opt.numIdenticalPairs([1,2,3,1,1,3])}")
```

## 3. Alternative Optimized Approach: Using `collections.Counter`
`Counter` is a subclass of the built-in dictionary, specifically designed to count hashable objects. We can use it to count frequencies, and then apply the mathematical combination formula $n(n-1) / 2$ to find the number of pairs.

```python
from collections import Counter

class SolutionCounter:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        # Counter creates a dictionary of frequencies, e.g., {1: 3, 2: 1, 3: 2}
        counts = Counter(nums)
        num_good_pairs = 0

        for count in counts.values():
            # Apply the combination formula for each frequency
            num_good_pairs += (count * (count - 1)) // 2
            
        return num_good_pairs

sol_counter = SolutionCounter()
print(f"Test 1: {sol_counter.numIdenticalPairs([1,2,3,1,1,3])}")
```

### Bonus: Exploring `collections.Counter`
Here are some examples of what makes `Counter` so powerful.

```python
# 1. Creating a Counter
word_list = ['apple', 'banana', 'apple', 'cherry', 'apple', 'banana']
count_list = Counter(word_list)
print("From list:", count_list)

# 2. Missing keys return 0 (no KeyError!)
print("Count for 'orange':", count_list['orange'])

# 3. Most common elements
print("Top 2 most common:", count_list.most_common(2))

# 4. Math Operations
c1 = Counter(a=3, b=1)
c2 = Counter(a=1, b=2)
print("\nc1 + c2:", c1 + c2) # Addition
print("c1 - c2:", c1 - c2) # Subtraction
```
