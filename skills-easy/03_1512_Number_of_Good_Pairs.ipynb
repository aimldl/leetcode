# [1512. Number of Good Pairs](https://leetcode.com/problems/number-of-good-pairs/description/?envType=problem-list-v2&envId=d5titt6h)

Given an array of integers nums, return the number of good pairs.

A pair (i, j) is called good if nums[i] == nums[j] and i < j.



Example 1:
```
Input: nums = [1,2,3,1,1,3]
Output: 4
```
Explanation: There are 4 good pairs (0,3), (0,4), (3,4), (2,5) 0-indexed.

Example 2:
```
Input: nums = [1,1,1,1]
Output: 6
```
Explanation: Each pair in the array are good.

Example 3:
```
Input: nums = [1,2,3]
Output: 0
```

Constraints:
```
1 <= nums.length <= 100
1 <= nums[i] <= 100
```

````python
def numIdenticalPairs(nums):
    num_good_pairs = 0
    length = len(nums)
    for i in range(length):
        for j in range(i+1, length):
            if nums[i] == nums[j] and i < j:
                num_good_pairs += 1

    return num_good_pairs
````

```Python3
from typing import List

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        counts = {}
        num_good_pairs = 0

        for num in nums:
            if num in counts:
                # Add the number of times we've seen this 'num' before
                # because it forms a new pair with each of them.
                num_good_pairs += counts[num]
                counts[num] += 1
            else:
                # First time seeing this number
                counts[num] = 1
                
        return num_good_pairs
 ```

````python
def numIdenticalPairsWithHashTable1(nums):
    # Example: nums = [1,2,3,1,1,3]

    counts = {}
    num_good_pairs = 0

    for num in nums:
        if num in counts:
            num_good_pairs += counts[num]
            counts[num] +=1
        else:
            # First time seeing this number
            counts[num] = 1

    return num_good_pairs
````

### Why this works?
If you have seen the number 1 three times already, and you find a fourth 1, that new 1 can pair up with the 1st, 2nd, and 3rd ones. So, you simply add 3 to your total pairs, and then update your count of 1s to 4.

```Python3
from typing import List
from collections import Counter

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        # Counter creates a dictionary of frequencies, e.g., {1: 3, 2: 1, 3: 2}
        counts = Counter(nums)
        num_good_pairs = 0

        for count in counts.values():
            # Apply the combination formula for each frequency
            num_good_pairs += (count * (count - 1)) // 2
            
        return num_good_pairs
```
