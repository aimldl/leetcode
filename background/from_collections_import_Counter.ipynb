# `from collections import Counter`

Counter is one of the most useful tools in Python's standard library. It is a subclass of the built-in dictionary (dict), specifically designed to count hashable objects quickly and cleanly.

Instead of writing manual loops to check if a key exists and then incrementing its value (which requires if/else statements or defaultdict), Counter does it all automatically.

Here is your guide to how it works, what makes it special, and where to find the official documentation.

## 1. Creating a Counter
You can initialize a Counter in several ways. The most common is passing it an iterable (like a list or a string).

```python
from collections import Counter

# 1. From a list
word_list = ['apple', 'banana', 'apple', 'cherry', 'apple', 'banana']
count_list = Counter(word_list)
print(count_list)
# Output: Counter({'apple': 3, 'banana': 2, 'cherry': 1})

# 2. From a string (counts characters)
count_str = Counter("abracadabra")
print(count_str)
# Output: Counter({'a': 5, 'b': 2, 'r': 2, 'c': 1, 'd': 1})

# 3. From a dictionary
count_dict = Counter({'a': 4, 'b': 2})

# 4. From keyword arguments
count_kwargs = Counter(cats=4, dogs=8)
```

## 2. What Makes Counter Special?
Because it is designed strictly for tallying, it comes with a few superpowers that regular dictionaries don't have.

### Missing Keys Return Zero
If you ask a regular dictionary for a key that doesn't exist, it throws a KeyError. A Counter simply returns 0.

```python
c = Counter(['apple', 'banana'])
print(c['cherry'])  # Output: 0
```

### Specialized Methods
Counter introduces a few built-in methods that make data analysis incredibly fast.

| Method | What it does | Example |
| :--- | :--- | :--- |
| `most_common(n)` | Returns a list of the top *n* elements and their counts. | `c.most_common(2)` → `[('a', 5), ('b', 2)]` |
| `elements()` | Returns an iterator repeating each element by its count. | `list(Counter(a=2, b=1).elements())` → `['a', 'a', 'b']` |
| `total()` | Returns the sum of all counts (Python 3.10+). | `Counter(a=2, b=3).total()` → `5` |
| `subtract()` | Subtracts counts provided by another iterable/mapping. | `c.subtract({'a': 1})` |

```python
from collections import Counter

class SolutionCounter:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        # Counter creates a dictionary of frequencies, e.g., {1: 3, 2: 1, 3: 2}
        counts = Counter(nums)
        num_good_pairs = 0

        for count in counts.values():
            # Apply the combination formula for each frequency
            num_good_pairs += (count * (count - 1)) // 2

        return num_good_pairs

sol_counter = SolutionCounter()
print(f"Test 1: {sol_counter.numIdenticalPairs([1,2,3,1,1,3])}")
```

### Math Operations
You can perform mathematical operations directly on Counter objects to combine or compare tallies.

```python
c1 = Counter(a=3, b=1)
c2 = Counter(a=1, b=2)

print(c1 + c2)  # Add counts: Counter({'a': 4, 'b': 3})
print(c1 - c2)  # Subtract counts: Counter({'a': 2}) (b is dropped because 1-2 <= 0)
print(c1 & c2)  # Intersection (min of both): Counter({'a': 1, 'b': 1})
print(c1 | c2)  # Union (max of both): Counter({'a': 3, 'b': 2})
```
